20/01/2569 เขียนโค้ดโปรแกรมการคำนวณเวลาอ่านหนังสือ

```python
name=input("ยินดีต้อนรับเข้าสู่โปรแกมคำนวณเวลาอ่านหนังสือ ผู้ใช้ชื่ออะไรค่ะ\n")

Reading1=int(input("กรุณาใส่เวลาว่างหรือเลิกเรียนแล้ว"))
Reading2=int(input("กรุณาใส่เลข24(หรือเวลาทั้งหมด1วัน)"))

total = Reading2 - Reading1

print("\nรวมเวลาการอ่านหนังสือของผู้ใช้:", total)

if total>=2:
  print("เวลาของคุณเหมาะสมกับการอ่านหนังสือ")

elif total<= 2:
  print("อ่านหนังสือของคุณไม่เหมาะสมเนื่องจากเวลาพักผ่อนน้อยเกินไป")

else:
  print("การคำนวณของคุณสำเร็จแล้ว")
```
