```python
#Ex3 BMI ver 1.0 ของ เลขที่ 30 ชื่อจริง บุญสิตา ธรรมเกษร ชื่อเล่นใบบัว

weight= float(input("ใส่น้ำหนักของคุณ(กก.) "))
height = float(input("ใส่ส่วนสูงของคุณ(ม.) "))
bmi = weight / (height**2)

if bmi < 18.5:
  print("ค่า BMI ของคุณ คือ ", bmi)
  print("แปลผล คือ น้ำหนักน้อย")
  print("คำแนะนำ คือ ทานอาหารให้ครบ 5 หมู่")
elif bmi< 23.0:
    print("ค่า BMI ของคุณ คือ", bmi)
    print("แปลผล คือ น้ำหนักปกติ")
    print("คำแนะนำ คือ รักษาสุขภาพด้วยนะเคิ้บ")
else:
      print("ค่า BMI ของคุณ คือ ", bmi)
      print("แปลผล คือ น้ำหนักเกิน")
      print("คำแนะนำ คือ มาออกกำลังกายกันเถอะ")
```

```python
#สอบกลางภาค 1
import math

name=input("ยินดีต้อนรับค่ะ! ผู้ใช้ชื่ออะไรค่ะ")
print("สวัสดีค่ะ,"+name)
print("\n โปรแกรมคำนวณพีทาโกรัส")

A= int(input('ใส่ค่าของด้านประกอบมุมฉาก1(ซม.)'))
B= int(input('ใส่ค่าของด้านประกอบมุมฉาก2(ซม.)'))
area = math.sqrt(A**2 + B**2)
print("พีทาโกรัส = ",area,"ซม.")
print("\n ขอบคุณค่า!")
print("\n เขียนโปรแกรมโดย : เลขที่30")
print("\n ชื่อเล่น : ใบบัว")
print("\n Good luck นะค้า!")
```

```python
#สอบกลางภาคเทอม 1
name=input("ยินดีต้อนรับค่า! ผู้ใชชื่ออะไรค่า")
print("สวัสดีเคิ้บบ,"+name)
print("ยินดีต้อนรับเข้าสู่โปรแกรมตรวจสอบคะแนนด้านการเรียนของนักเรียน\n")

dis = int(input("คะแนนความมีวินัย(เต็ม 100):"))
uni = int(input("คะแนนการส่งงานให้ตรงเวลาที่กำหนด(เต็ม 100)"))
politeness = int(input("คะแนนความสุภาพในการสื่อสารอย่างสุภาพ(เต็ม 100"))

total = dis + uni + politeness
aver = total/3

print("\nคะแนนรวมของนักเรียนคือ:", total)
print("คะแนนเฉลี่ย: %.2f" % aver)

print("\nสรุปผลด้านการเรียนของนักเรียน:")
if aver >=90:
  print("นักเรียนตัวอย่าง! ยอดเยี่ยมม")
elif aver >=70:
  print("มีด้านการเรียนที่ดี พอใช้ แต่ยังมีจุดที่ต้องพัฒนาอีกนิดนึงนะค่า")
elif aver >=50:
  print("มีด้านการเรียนที่พอใช้ ควรปรับปรุงหลายอย่าง")
else:
  print("มีด้านการเรียนที่ต่ำมาก อาจมีผลต่อเกรดเฉลี่ยด้านการเรียนในทางที่ไม่ดี")

print("\nขอบคุณค่า!")
print("เขียนโปรแกรมโดย : เลขที่ 30")
print("ชื่อเล่น : ใบบัว")
print("Good luck น้าค้า!")
```
